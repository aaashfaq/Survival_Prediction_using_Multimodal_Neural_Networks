```python
import os
import glob
import pandas as pd

def getSubstringBetweenTwoChars(char1, char2, input_str):
    start = input_str.find(char1) + 1
    end = input_str.find(char2)
    return input_str[start:end]

datasets = ['/root/code/thesis/codeFolder/LatestDataInUse/Combining/e-thrive_', '/root/code/thesis/codeFolder/LatestDataInUse/Combining/s3t1_wats_neu_Augmented', '/root/code/thesis/codeFolder/LatestDataInUse/Combining/t2_hr_spir_range_Augmented']

final_df = pd.DataFrame(columns=['PatientID', 'DataSource1', 'DataSource2', 'DataSource3', 'sCD25(IL-2Ra)', '4-1BB', 'CTLA-4', 'PD-L1', 'PD-1', 'Tim-3', '365'])

for dataset in datasets:
    df = pd.DataFrame(columns=['PatientID', 'Images', 'sCD25(IL-2Ra)', '4-1BB', 'CTLA-4', 'PD-L1', 'PD-1', 'Tim-3', '365'])
    
    for patient_id in os.listdir(dataset):
        bar_code = patient_id.split('-')[0].upper()

        data = excel.loc[excel['Bar-code'] == bar_code]

        sCD25 = data['sCD25   (IL-2Ra) (A4)'].tolist()
        sCD25 = str(sCD25)
        sCD25 = getSubstringBetweenTwoChars('[', ']', sCD25)

        bb14 = data['4-1BB  (A6)'].tolist()
        bb14 = str(bb14)
        bb14 = getSubstringBetweenTwoChars('[', ']', bb14)

        CTLA = data['CTLA-4 (B2)'].tolist()
        CTLA = str(CTLA)
        CTLA = getSubstringBetweenTwoChars('[', ']', CTLA)

        PDL1 = data['PD-L1 (B3)'].tolist()
        PDL1 = str(PDL1)
        PDL1 = getSubstringBetweenTwoChars('[', ']', PDL1)

        PD1 = data['PD-1    (B5)'].tolist()
        PD1 = str(PD1)
        PD1 = getSubstringBetweenTwoChars('[', ']', PD1)

        Tim3 = data['Tim-3   (B6)'].tolist()
        Tim3 = str(Tim3)
        Tim3 = getSubstringBetweenTwoChars('[', ']', Tim3)

        label365 = data['365'].tolist()
        label365 = str(label365)
        label365 = getSubstringBetweenTwoChars('[', ']', label365)

        for file in glob.glob(os.path.join(dataset, patient_id) + '**/**'):
            df = df.append({'PatientID': patient_id, 'Images': file, 'sCD25(IL-2Ra)': sCD25,
                            '4-1BB': bb14, 'CTLA-4': CTLA, 'PD-L1': PDL1, 'PD-1': PD1,
                            'Tim-3': Tim3, '365': label365}, ignore_index=True)

    # Rename columns to include the dataset name
    df.rename(columns={'Images': f'DataSource{datasets.index(dataset) + 1}'}, inplace=True)

    # Merge dataframes based on 'PatientID'
    if final_df.empty:
        final_df = df
    else:
        final_df = pd.merge(final_df, df, on='PatientID', how='outer')

# Fill NaN values with an appropriate default value if needed
final_df.fillna(value='', inplace=True)

print(final_df)
```
