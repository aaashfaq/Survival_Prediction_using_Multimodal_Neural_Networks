```python
import os
import glob
import zipfile
from pathlib import Path
import pandas as pd

import numpy as np
import pydicom
from PIL import Image
```

```python
Dataset='/root/code/thesis/codeFolder/LatestDataInUse/e-thrive_bh'
```

```python
excel=pd.read_csv('/root/Data/MRTData/MRTexcelDataTumorsize.csv')
```

```python
uniqueID=excel['Bar-code'].unique()
```

```python
len(uniqueID)
```

```python
df = pd.DataFrame(columns=['PatientID','Images','sCD25(IL-2Ra)','4-1BB', 'CTLA-4', 'PD-L1', 'PD-1','Tim-3', '365'])
#df = pd.DataFrame(columns=['PatientID','Images', 'Liver','Tumor','sCD25(IL-2Ra)','4-1BB', 'CTLA-4', 'PD-L1', 'PD-1','Tim-3', '365'])
```

```python
def getSubstringBetweenTwoChars(ch1,ch2,s):
    return s[s.find(ch1)+1:s.find(ch2)]
 
```

```python
for patient_id in os.listdir(Dataset):
    bar_code=patient_id.split('-')[0]
    bar_code=bar_code.upper()
    
    

    data=excel.loc[excel['Bar-code'] == bar_code]


    liver=data['size total liver [cm]'].tolist()
    liver=str(liver)
    liver= getSubstringBetweenTwoChars('[',']',liver)

    tumor=data['tumor size [cm]'].tolist()
    tumor=str(tumor)
    tumor= getSubstringBetweenTwoChars('[',']',tumor)
    
        
    sCD25=data['sCD25   (IL-2Ra) (A4)'].tolist()
    sCD25=str(sCD25)
    sCD25= getSubstringBetweenTwoChars('[',']',sCD25)
    
    bb14=data['4-1BB  (A6)'].tolist()
    bb14=str(bb14)
    bb14= getSubstringBetweenTwoChars('[',']',bb14)
    
    CTLA=data['CTLA-4 (B2)'].tolist()
    CTLA=str(CTLA)
    CTLA= getSubstringBetweenTwoChars('[',']',CTLA)
    
    PDL1=data['PD-L1 (B3)'].tolist()
    PDL1=str(PDL1)
    PDL1= getSubstringBetweenTwoChars('[',']',PDL1)
    
    PD1=data['PD-1    (B5)'].tolist()
    PD1=str(PD1)
    PD1= getSubstringBetweenTwoChars('[',']',PD1)
    
    
    Tim3=data['Tim-3   (B6)'].tolist()
    Tim3=str(Tim3)
    Tim3= getSubstringBetweenTwoChars('[',']',Tim3)
    
    label365=data['365'].tolist()
    label365=str(label365)
    label365= getSubstringBetweenTwoChars('[',']',label365)
    
    
    for file in glob.glob(os.path.join(Dataset,patient_id)+'**/**'):
        df = df.append({'PatientID':patient_id,'Images':file, 'sCD25(IL-2Ra)':sCD25,
                        '4-1BB':bb14, 'CTLA-4':CTLA, 'PD-L1':PDL1, 'PD-1':PD1,
                        'Tim-3':Tim3, '365':label365}, ignore_index=True)
            
        # df = df.append({'PatientID':patient_id,'Images':file,'Liver':liver,'Tumor':tumor, 'sCD25(IL-2Ra)':sCD25,
        #                 '4-1BB':bb14, 'CTLA-4':CTLA, 'PD-L1':PDL1, 'PD-1':PD1,
        #                 'Tim-3':Tim3, '365':label365}, ignore_index=True)
```
